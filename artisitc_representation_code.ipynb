La idea es poner todo el codigo de una manera artistica para que se vea mejor cuanto codigo se ha escrito.

Primero tenemos que tener una imagen con muy buena resolución, luego debemos de poner el texto en un rectangulo de una forma similar, y luego con las proporciones del rectangulo y los colores de los pixeles en dichas proporciones pondremos color a las letras.

```python
from PIL import Image, ImageDraw, ImageFont
import numpy as np 
import matplotlib.pyplot as plt 
import os 
import requests
!pip install lorem
import lorem
```

```python
limit=5e6
with open("/content/drive/MyDrive/proyectos_propios/print_image_with_code/calorro.txt","r") as fil:
    tex=fil.read()
#texto=lorem.text()
tex=tex.replace("\t"," ").replace("\n"," ")
texto=""
while len(texto)<limit:
    texto=texto+tex#+lorem.text()

len(texto)
```

```python
texto[:450]
```

```python
# get an image
url="https://i0.wp.com/plumasatomicas.com/wp-content/uploads/2020/04/Ultima-Cena-Jueves-Santo-Cuadro-Da-Vinci.jpg?resize=1200%2C720&ssl=1"
url2="https://tcf.admeen.org/game/15000/14518/400x246/red-ball-4-volume-1.jpg"
img = Image.open(requests.get(url,stream=True).raw).convert("RGB")
#r,g,b=img.split()
img_np=np.asarray(img)
#font = ImageFont.truetype("/content/drive/MyDrive/proyectos_propios/print_image_with_code/Traveling__Typewriter.ttf", 12)
img_text=Image.new("RGB",tuple(np.asarray(img.size)*7),(255,255,255))
#img_text=Image.new("RGB",(img.size[0]*6,img.size[1]*7),(255,255,255))
draw = ImageDraw.Draw(img_text)
```

```python
for i in range(img.size[0]):
    for j in range(img.size[1]):
        draw.text((i*7,j*7),texto[j*img.size[0]+i],fill=tuple(img_np[j,i]))

```

```python
img.size[0]*img.size[1]
```

```python
center=[img_text.size[0]//2,img_text.size[1]//2]
```

```python
img_crop=img_text.crop((center[0]-400,center[1]-700,center[0]+200,center[1]+100))
```

```python
plt.figure(figsize=(13,13))
plt.imshow(img_text)
plt.xticks([])
plt.yticks([])
```

```python
plt.figure(figsize=(13,13))
plt.imshow(img_crop)
plt.xticks([])
plt.yticks([])
```
